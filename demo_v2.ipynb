### Integer Linear Programing 적용에 필요한 함수

```python
from datetime import date, datetime, timedelta

# 사용자 일정(구글 캘린더, 수면) 전처리 함수
def preprocessing_schedules(week_start, week_end, raw_schedules):
    fixed_schedules = []
    for item in raw_schedules:
        start_dt = item['start']['dateTime']
        end_dt = item['end']['dateTime']

        # +09:00 제거
        start_dt = datetime.fromisoformat(start_dt[:-6])
        end_dt = datetime.fromisoformat(end_dt[:-6])

        # 여러 날에 걸친 일정을 하루 단위로 분리
        current = start_dt
        while current.date() <= end_dt.date():
            day_start = current.replace(hour=0, minute=0, second=0, microsecond=0)
            day_end = day_start + timedelta(days=1)

            seg_start = max(start_dt, day_start)
            seg_end = min(end_dt, day_end)

            if not (week_start <= seg_start.date() < week_end): # 이번주 데이터가 아닌 부분은 버린다
                current = day_end
                continue

            weekday = seg_start.weekday() # 요일 0=월, 6=일
            start_hour = seg_start.hour + seg_start.minute / 60
            end_hour = seg_end.hour + seg_end.minute / 60

            fixed_schedules.append({
                'summary': item['summary'],
                'location': item['location'],
                'day': weekday,
                'start': start_hour,
                'end': end_hour
            })

            current = day_end # 다음 날짜

    return fixed_schedules
```

```python
import pulp

def generate_initial_schedule(cleaning_tasks, user_cleaning_status, fixed_schedules,
                              total_cleaning_task_limit=10, total_cleaning_time_limit=5):
    # 1) ILP 문제 정의
    prob = pulp.LpProblem("CleaningScheduler", pulp.LpMaximize)

    # 2) 변수: 각 청소를 x요일 x시에 할지 말지 정함
    variables = {}
    for i, task in enumerate(cleaning_tasks):
        for d in range(7):  # 0 ~ 6(월요일~일요일)
            for h in range(24):  # 청소 시작시간
                variables[(i, d, h)] = pulp.LpVariable(f'task_{i}_day{d}_hour{h}', 0, 1, cat='Binary')

    # 3) 목적 함수 : 'interval', 'last_done_days_ago' 만족 -> 우선순위가 높은(미룬지 오래된) 청소 작업을 더 많이 포함
    objective = pulp.lpSum(
        (user_cleaning_status[task_name]["last_done_days_ago"] - task['interval']) * variables[(i, d, h)]
        for i, task_dict in enumerate(cleaning_tasks)
        for task_name, task in task_dict.items()
        for d in range(7) for h in range(24)
    )
    prob += objective

    # 4) 제약 조건
    # 제약 조건 안에서 최대한 많은 청소를 배정
    prob += pulp.lpSum([
        variables[(i, d, h)]
        for i in range(len(cleaning_tasks))
        for d in range(7) for h in range(24)])

    # 총 청소 개수 제한
    prob += pulp.lpSum([
        variables[(i, d, h)]
        for i in range(len(cleaning_tasks))
        for d in range(7) for h in range(24)]) <= total_cleaning_task_limit

    # 총 청소 시간 제한
    prob += pulp.lpSum([
        variables[(i, d, h)] * task['duration']
        for i, task_dict in enumerate(cleaning_tasks)
        for task_name, task in task_dict.items()
        for d in range(7) for h in range(24)]) <= total_cleaning_time_limit

    # 고정 일정 제외한 시간에만 청소 배정
    for i, task_dict in enumerate(cleaning_tasks):
        for task_name, task in task_dict.items():
            duration = task["duration"]
            for d in range(7):
                for h in range(24):
                    for sched in fixed_schedules:
                        if sched['day'] == d:
                            if sched['start'] < h + duration and sched['end'] > h:
                                prob += variables[(i, d, h)] == 0

    # 같은 시간대(x시:00분)에 실행되는 청소의 duration 합 <= 1시간
    for d in range(7):
        for h in range(24):
            prob += pulp.lpSum([
                variables[(i, d, h)] * task["duration"]
                for i, task_dict in enumerate(cleaning_tasks)
                for task in task_dict.values()
            ]) <= 1

    # 각 청소작업은 1주일에 (7 // interval)회 이하만 추천
    for i, task_dict in enumerate(cleaning_tasks):
        for task_name, task in task_dict.items():
            max_assign = 7 // task["interval"]
            prob += pulp.lpSum([
                variables[(i, d, h)]
                for d in range(7) for h in range(24)
            ]) <= max_assign

    # 최적화 수행
    prob.solve()

    # 결과 추출
    cleaning_schedules = []
    for i, task_dict in enumerate(cleaning_tasks):
        for task_name, task in task_dict.items():
            for d in range(7):
                for h in range(24):
                    if pulp.value(variables[(i, d, h)]) == 1:
                        cleaning_schedules.append([
                            task_dict, d, h
                        ])

    return cleaning_schedules
```

### 직전 주차 데이터 처리에 필요한 함수

```python
# user_cleaning_status의 'last_done_days_ago' 업데이트
def compute_user_cleaning_status(user_execution_data):
    global user_cleaning_status

    not_done = set(list(user_cleaning_status.keys())) # 최종적으로 안 한 청소만 남기기 위함
    for entry in user_execution_data:
        if not entry['performed']:
            continue
        task = entry['task']
        prev_days_ago = user_cleaning_status[task]['last_done_days_ago']
        new_days_ago = 7 - entry['day'] # 며칠 전에 했는지 계산
        if new_days_ago < prev_days_ago: # 1주 동안 같은 task를 여러번 수행했을 경우 가장 최신 정보를 반영함
            user_cleaning_status[task]['last_done_days_ago'] = new_days_ago
        not_done.discard(task) # 수행한 청소는 삭제

    for task in not_done:
        user_cleaning_status[task]['last_done_days_ago'] += 7 # 이번주에 안 한 청소는 'last_done_days_ago' 7일 증가
```

```python
import numpy as np

# 시간 슬롯 당 청소 확률 벡터 계산(behavior vector)
def compute_user_behavior(user_execution_data, prev_behavior=None, alpha=0.5):
    '''
    Parameters:
    - user_execution_data: 지난주 청소 기록
    - prev_behavior: shape=(168,), 누적된 행동 벡터
    - alpha: 이전 행동 벡터 반영 비율

    Returns:
    - behavior_vector: shape=(168,), np.float32
    '''
    vector = np.zeros(168)
    for entry in user_execution_data:
        index = entry['day'] * 24 + entry['hour'] # 168개 중 하나의 인덱스로 계산
        if entry['performed']:
            vector[index] += 1.0

    # 0.0 ~ 1.0 사이로 정규화 한 값 저장
    if vector.sum() > 0:
        vector /= vector.sum()
    vector = vector.astype(np.float32)

    if prev_behavior is None:
        return vector
    else:
        # EWMA 업데이트
        return alpha * prev_behavior + (1 - alpha) * vector
```

```python
# 직전 주의 청소 완료 기록 데이터만 추출
def prev_execution(user_execution_data, this_week=2):
    data = [entry for entry in user_execution_data if entry['week'] == this_week - 1]
    return data
```

### RL 준비 코드

```python
# **additional update with RL**
import numpy as np
import gymnasium as gym
from gymnasium import spaces
import torch
import torch.nn as nn
import torch.optim as optim
from tianshou.env import DummyVectorEnv
from tianshou.policy import PPOPolicy
from tianshou.data import Collector, VectorReplayBuffer
from tianshou.trainer import onpolicy_trainer
from tianshou.utils.net.common import Net
from torch.distributions import Categorical
from torch.nn.utils import clip_grad_norm_

# 환경 정의
class CleaningEnv(gym.Env):
    def __init__(self, fixed_schedules, cleaning_tasks, user_cleaning_status, user_behavior, user_execution_data):
        super().__init__()

        self.fixed_schedules = fixed_schedules
        self.cleaning_tasks = cleaning_tasks
        self.user_cleaning_status = user_cleaning_status
        self.behavior_vector = user_behavior # 길이 168, 시간 선호도, 값이 높을수록 사용자가 해당 시간에 청소할 확률이 높음
        self.user_execution_data = user_execution_data

        self.n_tasks = len(cleaning_tasks)
        self.current_step = 0
        self.max_steps = 10 # 수정 가능

        # 요일 0~6, 시간대 0~23 -> 각 자리의 범위는 7, 24 미만
        self.action_space = spaces.MultiDiscrete([7, 24] * self.n_tasks)

        # 24x7 = 168개의 시간 슬롯, 각 슬롯마다 (fixed_vector, urgency_vector, behavior_vector) 존재,
        self.observation_space = spaces.Box(low=0.0, high=1.0, shape=(168 * 3,), dtype=np.float32)
            # fixed_vector: 사용자의 고정 일정
            # urgency_vector: 청소 급함도(0, 1사이 실수)
            # behavior_vector: 시간대 별 상대적인 선호도

        self.state = None


    def reset(self, *, seed=None, options=None):
        super().reset(seed=seed)
        self.current_step = 0
        self.state = self._get_initial_state()

        # 사용자 수행 데이터 반영 (self.state에 수행 완료 표시)
        for entry in self.user_execution_data:
            if entry['performed']:
                idx = int(entry['day'] * 24 + entry['hour'])
                self.state[168 + idx] = 1

        return self.state, {} # state, (obs, info)

    def _get_initial_state(self):
        fixed_vector = np.zeros(168)
        urgency_vector = np.zeros(168)

        # 각 벡터에 정보 저장
        for sched in self.fixed_schedules:
            start = int(sched['day'] * 24 + sched['start'])
            end = int(sched['day'] * 24 + sched['end'])
            fixed_vector[start:end] = 1 # 고정일정이 있는 시간을 1로 표시

        for task in self.cleaning_tasks:
            name = list(task.keys())[0]
            info = task[name]
            if info['interval'] == 0:
                urgency = 1.0
            urgency = min(1.0, user_cleaning_status[name]['last_done_days_ago'] / (info['interval'] + 1e-8))
            # 맨 처음 urgency는 편향되지 않은 학습을 위해 분산시킨다.
            urgency_vector += urgency / 168.0

        urgency_vector = np.clip(urgency_vector, 0, 1) # 클리핑

        state = np.concatenate([fixed_vector, urgency_vector, self.behavior_vector], axis=0).astype(np.float32)
        if np.isnan(state).any() or np.isinf(state).any():
            print("Warning: state contains NaN or Inf!")
            print(state)
        return state


    def step(self, action): # action = [day1, time_slot1, day2, time_slot2, ..., dayN, time_slotN], len(action) = 2 * n_tasks
        cleaning_schedules = []

        for i in range(self.n_tasks):
            task = self.cleaning_tasks[i]
            day = action[2 * i]
            hour_slot = action[2 * i + 1]
            cleaning_schedules.append((task, day, hour_slot))

            # update state
            idx = int(day * 24 + hour_slot)
            self.state[168 + idx] = 0.0 # urgency_vector 0으로 설정(이제는 급한 청소가 아님)

        reward = self._calculate_reward(cleaning_schedules)

        for entry in self.user_execution_data:
            if entry['performed']:
                match = any(
                    t[0] == entry['task'] and t[1] == entry['day'] and t[2] == entry['hour_slot']
                    for t in cleaning_schedules
                )
                reward += 1.0 if match else -1.0 # 실제 수행한 작업과 일치하는 스케줄을 더 선호하도록 보상을 조정

        self.current_step += 1
        terminated = self.current_step >= self.max_steps
        truncated = False
        next_state = self.state.copy()

        return next_state, reward, terminated, truncated, {}

    def _calculate_reward(self, schedules):
        cleanliness_score = 0
        burden_penalty = 0
        conflict_penalty = 0
        preference_score = 0

        for task_tuple in schedules: # schedules: 초기 청소 일정(cleaning_schedules)
            task, day, start = task_tuple
            name = list(task.keys())[0] # 청소 이름
            info = task[name] # 해당 청소의 정보
            idx = int(day * 24 + start) # 168개 시간 슬롯에 대한 index

            # 고정 일정과 충돌 확인 -> RL 업데이트 시 고정 일정과 충돌하면 패널티 줌.
            for sched in self.fixed_schedules:
                if sched['day'] == day and sched['start'] <= start < sched['end']:
                    conflict_penalty += 1
                    break

            # 늦은 시간, 새벽 부담
            if start < 7 or start > 22:
                burden_penalty += 0.5

            # 청소 필요도(급한 청소)
            overdue = user_cleaning_status[name]['last_done_days_ago'] - info['interval']
            if overdue >= 0:
                cleanliness_score += 1.0
            else:
                cleanliness_score += 0.5

            # 사용자 시간 선호도 - user_behavior(behavior_vector)
            preference_score += self.behavior_vector[idx] * 2.0 # 가중치 변경 가능

        reward = float(cleanliness_score - burden_penalty - conflict_penalty * 2.0 + preference_score)


        return reward

# 필요한 함수, wrapper 클래스

def to_tensor(x): # 다양한 형태의 x를 텐서로 변환
    if isinstance(x, torch.Tensor):
        return x
    elif isinstance(x, np.ndarray):
        return torch.from_numpy(x).float()
    elif isinstance(x, (list, tuple)):
        return torch.tensor(np.array(x), dtype=torch.float32) #
    elif isinstance(x, (float, int, np.float32, np.float64, np.int32, np.int64)):
        return torch.tensor([x], dtype=torch.float32)
    else:
        raise TypeError(f"Unsupported input type: {type(x)}")

class MultiHead(nn.Module):
    def __init__(self, input_dim, nvec):
        super().__init__()
        self.heads = nn.ModuleList([nn.Linear(input_dim, n) for n in nvec])

    def forward(self, x=None, state=None, **kwargs):
        x = to_tensor(x)
        return [head(x) for head in self.heads]

class ActorNet(nn.Module):
    def __init__(self, net: nn.Module, action_nvec):
        super().__init__()
        self.backbone = net
        self.head = MultiHead(net.output_dim, action_nvec)

    def forward(self, x, state=None, info=None):
        x = to_tensor(x) # [4, 504]

        if torch.isnan(x).any() or torch.isinf(x).any():
            raise ValueError(f"Input to Backbone contains NaN or Inf: {x}")

        x, _ = self.backbone(x)

        if torch.isnan(x).any() or torch.isinf(x).any():
            raise ValueError(f"Backbone output contains NaN or Inf: {x}")

        logits_list = self.head(x)

        for i, logits in enumerate(logits_list):
            if torch.isnan(logits).any() or torch.isinf(logits).any():
                raise ValueError(f"Head output logits_{i} contains NaN or Inf: {logits}")

        logits = {f'logits_{i}': l for i, l in enumerate(logits_list)}

        return logits, state

class CriticWrapper(nn.Module):
    def __init__(self, critic):
        super().__init__()
        self.critic = critic

    def forward(self, x):
        x = x[0] # (obs, state) 중 obs만 사용
        x = to_tensor(x)
        if x.dim() < 2:
            x = x.unsqueeze(0)

        out = self.critic[0](x)

        if isinstance(out, tuple):
            out = out[0]
            print(out)

        if torch.isnan(out).any() or torch.isinf(out).any():
            raise ValueError("Output contains NaN or Inf1") # nan 에러 발생

        out = self.critic[1](out)

        if torch.isnan(out).any() or torch.isinf(out).any():
            raise ValueError("Output contains NaN or Inf2")

        return out

class CustomPPOPolicy(PPOPolicy):
    def update(self, *args, **kwargs):

        result = super().update(*args, **kwargs) # nan 에러 발생

        # 그레이디언트 클리핑 추가
        torch.nn.utils.clip_grad_norm_(self.parameters(), max_norm=0.5)

        return result

class DistributionTuple:
    """Tianshou에서 다중 Categorical 분포 처리를 위한 커스텀 래퍼 클래스."""
    def __init__(self, distributions):
        self.distributions = distributions

    def sample(self):
        return torch.stack([dist.sample() for dist in self.distributions], dim=-1)

    def log_prob(self, actions):
        # actions: shape [batch_size, n_tasks]
        log_probs = [dist.log_prob(actions[:, i]) for i, dist in enumerate(self.distributions)]
        return torch.stack(log_probs, dim=-1).sum(dim=-1) # shape: [batch_size]

    def entropy(self):
        return sum([dist.entropy() for dist in self.distributions])

    def mode(self):
        return torch.stack([dist.probs.argmax(dim=-1) for dist in self.distributions], dim=-1)

# 환경 생성 함수
def make_env():
    return CleaningEnv(fixed_schedules, cleaning_tasks, user_cleaning_status, user_behavior, prev_execution(user_execution_data))

# 하이퍼파라미터
batch_size = 64
train_num = 8  # 병렬 env 개수
test_num = 4 # 병렬 테스트 환경 수
epoch = 10
step_per_epoch = 1000
max_episode_steps = 10

# 환경 생성
train_envs = DummyVectorEnv([make_env for _ in range(train_num)])
test_envs = DummyVectorEnv([make_env for _ in range(test_num)])

# 네트워크 정의
obs_shape = train_envs.workers[0].env.observation_space.shape or train_envs.workers[0].env.observation_space.n
action_space = train_envs.workers[0].env.action_space
action_nvec = action_space.nvec.tolist()

def make_net():
    return Net(obs_shape[0], hidden_sizes=[128, 128], activation=nn.ReLU)

actor = ActorNet(make_net(), action_nvec)
critic = nn.Sequential(make_net(), nn.Linear(128, 1))
critic = CriticWrapper(critic)

optimizer = optim.Adam(list(actor.parameters()) + list(critic.parameters()), lr=1e-4)

def dist_fn(logits_dict):
    # logits_dict: {logits_0: [B, 7], logits_1: [B, 24], ...}
    dists = [Categorical(logits=logits_dict[f'logits_{i}']) for i in range(len(logits_dict))]
    return DistributionTuple(dists)

def init_weights(m):
    if isinstance(m, torch.nn.Linear):
        torch.nn.init.xavier_uniform_(m.weight)
        if m.bias is not None:
            torch.nn.init.zeros_(m.bias)

policy = CustomPPOPolicy(
    actor, critic, optim=optimizer, dist_fn=dist_fn,
    discount_factor=0.99, max_grad_norm=0.5,
    eps_clip=0.2, gae_lambda=0.95,
    action_space=action_space,
    reward_normalization=True,
    recompute_advantage=True,
    max_batchsize=batch_size,
)

policy.apply(init_weights)

# Collector 생성
train_collector = Collector(policy, train_envs, VectorReplayBuffer(batch_size * train_num, train_num))
test_collector = Collector(policy, test_envs)

# 테스트 함수
def test_fn(epoch, env_step):
    print(f"Epoch {epoch} | Env steps {env_step}")
    test_collector.collect(n_episode=10, n_step=None)
```

### RL 학습

```python
# 학습 시작
result = onpolicy_trainer(
    policy,
    train_collector,
    test_collector,
    max_epoch=epoch,
    step_per_epoch=step_per_epoch,
    repeat_per_collect=4,
    batch_size=batch_size,
    episode_per_test=test_num,
    episode_per_collect=8,
    test_fn=test_fn,
)
print(f"Training finished. Best reward: {result['best_reward']}")
```
